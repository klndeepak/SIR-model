```python
# importing all required libraries
import scipy.integrate 
import numpy as np
import matplotlib.pyplot as plt
import pylab
```

```python
# Initializing all parameters
# N is the total population (1 Million).
N = 10000000
# Initial number of infected is I0 and recovered individuals is R0.
I0, R0 = 1, 0
# S0 is the population susceptible to infection initially.
S0 = N - I0 - R0
# Contact rate is denoted by beta, and mean recovery rate is gamma (denoted in 1/days).
beta, gamma = 0.4, 0.1 
beta1,gamma1 = 0.6, 0.1
```

```python
# A grid of time points (in days)
# linspace (start, stop, number of points in middle)
t = np.linspace(0, 160, 100)
```

```python
# The SIR model differential equations.
# declare a vector y for S, I, and R
def SIR_model(y, t, N, beta, gamma):
    S, I, R = y 
    dS_dt = -(beta * S * I / N)
    dI_dt = (beta * S * I / N) - (gamma * I)
    dR_dt = gamma * I
    return ([dS_dt, dI_dt, dR_dt])
```

```python
def SIR_model1(y1, t, N, beta1, gamma1):
    S, I, R = y1 
    dS1_dt = -(beta1 * S * I / N)
    dI1_dt = (beta1 * S * I / N) - (gamma1 * I)
    dR1_dt = gamma1 * I
    return ([dS1_dt, dI1_dt, dR1_dt])
```

```python
# Initial conditions vector y0
y0 = S0, I0, R0

solution = scipy.integrate.odeint(SIR_model,[S0, I0, R0], t, args=(N,beta,gamma))
solution = (0.0000001)*np.array(solution)
```

```python
# Initial conditions vector y0
y0 = S0, I0, R0

solution1 = scipy.integrate.odeint(SIR_model1, [S0,I0,R0],t,args=(N,beta1,gamma1))
solution1 = (0.0000001)*np.array(solution1)
```

```python
plt.figure(figsize = [6,6])
plt.plot(t, solution[:,0], label="S(t)")
plt.plot(t, solution[:,1], label="I(t)")
plt.plot(t, solution[:,2], label="R(t)")
plt.grid()
plt.legend(loc='best',fontsize=20)
plt.xlabel("Time (days)", fontname="Times New Roman",fontsize=20)
plt.ylabel("Population/Million",fontname="Times New Roman",fontsize=20)
plt.tick_params(labelsize=15, pad=6)
#plt.title("SIR model")
pylab.xlim([0,100])
pylab.ylim([0,1])
plt.show


plt.figure(figsize = [6,6])
plt.plot(t, solution1[:,0], label="S(t)")
plt.plot(t, solution1[:,1], label="I(t)")
plt.plot(t, solution1[:,2], label="R(t)")
plt.grid()
plt.legend(loc='right',fontsize=20)
plt.xlabel("Time (days)",fontname="Times New Roman",fontsize=20)
plt.ylabel("Population/Million",fontname="Times New Roman",fontsize=20)
plt.tick_params(labelsize=15, pad=6)
pylab.xlim([0,100])
pylab.ylim([0,1])
plt.show
```
